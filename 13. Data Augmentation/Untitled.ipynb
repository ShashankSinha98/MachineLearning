## Data Augmentation
<ul>
<li>Deep Learning Models are "Data Hungry"</li>
<li>Small Datasets leads to overfitting most of the time</li>
<li>So generate more data by augmentating the training samples</li></ul>

```python
img_path = "image.jpg"
```

```python
from keras.preprocessing import image
import matplotlib.pyplot as plt
```

```python
img = image.load_img(img_path,target_size=(500,500))
```

```python
x = image.img_to_array(img)
```

```python
plt.imshow(x)
plt.axis("off")
plt.show()
```

```python
x = x.reshape((1,500,500,3))
print(x.shape)
```

```python
datagen = image.ImageDataGenerator(
    rescale=1.0/255.0,
    rotation_range=30,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode="nearest"
)
```

```python
print(type(datagen))
```

```python
i=0
for batch in datagen.flow(x,batch_size=1):
    print(type(batch))
    print(batch.shape)
    imgplot = plt.imshow(image.img_to_array(batch[0]))
    plt.show()
    i+=1
    if i==10:
        break
        
        
```
