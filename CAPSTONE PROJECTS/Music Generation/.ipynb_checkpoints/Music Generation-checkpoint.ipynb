```python
from music21 import converter, instrument, note, chord, stream
import glob
import pickle
import numpy as np
```

## Read a Midi File

```python
# Loading Midi File from external dir
midi = converter.parse('midi_songs/Eternal_Harvest.mid')
```

```python
midi
```

```python
## Stream.Score is a container type class which contains both notes and chords
```

```python
# Playing Midi file
midi.show('midi')
```

```python
# Getting all notes, chords and offsets of midi file
midi.show('text')


## STRUCTURE:
## 1.Container(stream.Score)
##        |-  1.1 Subcontainer
##        |         |- Notes
##        |         |- Chords
##        |
##        |-  1.2 Subcontainer
##        |         |- Notes
##        |         |- Chords
##        |
##.. and so on
## Like List of List [[N,C..],[N,C,...]]
```

```python
# Flatten all list to get a single list consisting of Notes and chords only
elements_to_parse = midi.flat.notes # (midi.flat.notes-Iterator)
```

```python
# No of all notes and chords present- (<music21.chord.Chord D5 G4> - 1 element)
len(elements_to_parse)
```

```python
# Offset - time duration at which a particular note/chord is played
for e in elements_to_parse:
    print(e, e.offset)
```

## 2. Prparing Data for ML model
<ul>
<li>Read all midi files and store all notes and chords in separate single list</li>
</ul>

```python
# elements_to_parse contains both notes and chords. we need to process them as follows

# 1. Note Processing
print(elements_to_parse[3])
print(elements_to_parse[3].pitch) # Gonna store it as string

# 2. Chord Processing (multiple notes played at same time)
print(elements_to_parse[100])
print(elements_to_parse[100].normalOrder) # get in representation of all notes present int this chord
                                          # 0 - E-5, 3 - C4
# We'll concatenate all notes in chords with '+' and store them as string
concat_chord = '+'.join(str(n) for n in elements_to_parse[100].normalOrder)
print(concat_chord) #"0+3" means 2 notes played in this chord {0 - E-5, 3 - C4}
```
