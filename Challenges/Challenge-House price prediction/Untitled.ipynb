```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
```

```python
X_Train = pd.read_csv('Train_Data.csv')
```

```python
# Columns to drop - Id, Alley, PoolQC, Fence, MiscFeature
X_Train = X_Train.drop(['Id','Alley','PoolQC','Fence','MiscFeature','MasVnrArea'],axis=1)
```

```python
str_cols = []
for i,(colname, coltype) in enumerate(X_Train.dtypes.iteritems()):
    if coltype == "object":
        str_cols.append(i)
        
```

```python
X_Train = X_Train.fillna("nan")
print(str_cols)
print(len(str_cols))
```

```python
X_Train = X_Train.values
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
X_Train[:,2:3] = imputer.fit_transform(X_Train[:,2:3])
X_Train[:,56:57] = imputer.fit_transform(X_Train[:,56:57])
```

```python
onehotencoder = OneHotEncoder(categories=str_cols)
ct = ColumnTransformer(
    [('one_hot_encoder', OneHotEncoder(categories='auto'), str_cols)],   # The column numbers to be transformed (here is [0] but can be [0, 1, 3])
    remainder='passthrough'                                         # Leave the rest of the columns untouched
)
X_Train = ct.fit_transform(X_Train)
```

```python
X_Train = X_Train.toarray()
print(type(X_Train))
```
