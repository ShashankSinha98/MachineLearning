```python
# Loading Libraries
import cv2
import sys
import matplotlib.pyplot as plt
faceCascade = cv2.CascadeClassifier("haarcascade_frontalface_alt.xml")
```

```python
# Loading Images
face = cv2.cvtColor(cv2.imread('jamie.jpg'),cv2.COLOR_BGR2RGB)
glasses = cv2.cvtColor(cv2.imread('glasses.png'),cv2.COLOR_BGR2RGB)
mst = cv2.cvtColor(cv2.imread('mustache.png'),cv2.COLOR_BGR2RGB)
plt.imshow(mst)
```

```python
face_copy = face.copy()
faces = faceCascade.detectMultiScale(face)
if len(faces) > 0:
    print("Face DETECTED: ",faces)

    faces = sorted(faces, key=lambda f:f[2]*f[3])
    
    for x,y,w,h in faces[-1:]:
        cv2.rectangle(face_copy,(x,y),(x+w,y+h),(255,0,0),3)
    plt.imshow(face_copy)
    
else:
    print("Face NOT DETECTED!! ")
```

```python
face_copy = face.copy()
ly1 = int(y+0.65*h)
ry1 = int(ly1+0.1*h)
lx1 = int(x+0.25*w)
rx1 = int(lx1+0.48*w)
cv2.rectangle(face_copy,(lx1,ly1),(rx1,ry1),(255,0,0),3)
plt.imshow(face_copy)
```

```python
face_copy = face.copy()
ly2 = int(y+h/3)
lx2 = int(x+w/7)
ry2 = int(ly2+h/7)
rx2 = int(lx2+(3*w)/4)
cv2.rectangle(face_copy,(lx2,ly2),(rx2,ry2),(255,0,0),3)
plt.imshow(face_copy)
```

```python
face_copy = face.copy()
mst_width = int((rx1-lx1))
mst_height = int((ry1-ly1)*2)
glass_width = int((rx2-lx2)*1.3)
glass_height = int((ry2-ly2)*2.5)

mst = cv2.resize(mst,(mst_width,mst_height))
glasses = cv2.resize(glasses,(glass_width,glass_height))

for i in range(mst_height):
        for j in range(mst_width):
            for k in range(3):
                if mst[i][j][k]<235: #avoiding white
                    face_copy[int(ly1)+i][int(lx1)+j][k] = mst[i][j][k]
                    
                    
for i in range(glass_height):
    for j in range(glass_width):
        for k in range(3):
            if glasses[i][j][k]<235: #avoiding white
                face_copy[int(ly2*.9)+i][int(lx2*.9)+j][k] = glasses[i][j][k]
                
plt.imshow(face_copy)
```
