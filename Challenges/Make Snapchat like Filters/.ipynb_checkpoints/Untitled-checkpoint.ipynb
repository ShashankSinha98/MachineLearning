```python
# Loading Libraries
import cv2
import matplotlib.pyplot as plt
cascPath = "haarcascade_frontalface_alt.xml"
faceCascade = cv2.CascadeClassifier(cascPath)
eyeCascade = cv2.CascadeClassifier("haarcascade_eye.xml")
```

```python
# Loading Images
tyrion = cv2.cvtColor(cv2.imread('tyrion.png'),cv2.COLOR_BGR2RGB)
glasses = cv2.cvtColor(cv2.imread('glasses.png'),cv2.COLOR_BGR2RGB)
jamie = cv2.cvtColor(cv2.imread('Jamie_Before.jpg'),cv2.COLOR_BGR2RGB)
```

```python
# Visualizing - Jamie
jamie_face = faceCascade.detectMultiScale(jamie,1.3,1)
x,y,w,h = jamie_face[0]
print(x,y,w,h)
```

```python
jamie = cv2.cvtColor(cv2.imread('Jamie_Before.jpg'),cv2.COLOR_BGR2RGB)
cv2.rectangle(jamie,(217,385),(217+319,385+73),(255,0,0),3)
plt.imshow(jamie)
```

```python
# Visualizing - Tyrion
tyrion_face = eyeCascade.detectMultiScale(tyrion)
print(tyrion_face[0])
print(tyrion_face[1])
```

```python
tyrion = cv2.cvtColor(cv2.imread('tyrion.png'),cv2.COLOR_BGR2RGB)
glasses = cv2.cvtColor(cv2.imread('glasses.png'),cv2.COLOR_BGR2RGB)
glass_width = 151
glass_height = 51

glasses = cv2.resize(glasses,(glass_width,glass_height))

for i in range(glass_height):
    for j in range(glass_width):
        for k in range(3):
            if glasses[i][j][k]<235: #avoiding white
                tyrion[126+i][75+j][k] = glasses[i][j][k]
                
plt.imshow(tyrion)
```
