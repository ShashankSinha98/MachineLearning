```python
import gym
```

```python
env = gym.make('MountainCar-v0')
```

```python
env.reset()
```

```python
# Total no of actions available
print(env.action_space)
print(env.action_space.n)
print(env.observation_space)
print(env.action_space.sample()) # Randomly select one action
```

```python
env.reset()
   
for t in range(200):
    random_action = env.action_space.sample()
    env.step(random_action) # Perform action
    env.render()
    
env.close()
```

### 2. Playing multiple game episode with random strategy

```python
# Playing 2 game episode
for e in range(1):
    
    observation = env.reset()
    state = 0
    for t in range(200):
        
        if t%50==0:
            if state==0:
                state=2
            else:
                state=0
        
        env.render()
        random_action = env.action_space.sample()
    
        observation,reward,done,other_info = env.step(state) # Observation = new state
        print(observation,reward,done,other_info)
        if done:
            # Game episode over
            print("Game Episode %d/2, Score: %d"%(e+1,t))
            break
            
env.close()
print("All 20 episode over!")
        
```
